Este programa calcula la resta del impuseto segun el rango de su salario, con solo ingresar su salario

```
print("Descuento de impuestos")
x=float(input("Introduzca su salario neto: "))

if (x <= 10000):
  parcial = x * .05
  total = x - parcial

  print("Su salario esta en un rango menor a 10 MIL")
  print("Su salario total es: ",total)
  print("TRUE")

elif (x >= 10001 and x <= 15000):
  parcial = x * .10
  total = x - parcial

  print("Su salario esta en un rango entre 10 MIL y 15 MIL")
  print("Su salario total es: ",total)
  print("TRUE")

elif (x >= 15001 and x <= 20000):
  parcial = x * .15
  total = x - parcial

  print("Su salario esta en un rengo entre 15 MIL y 20 MIL")
  print("Su salario total es: ",total) 
  print("TRUE")

elif (x > 20001):
  parcial = x * .18
  total = x - parcial

  print("Su salario esta en un rango mayor a 20 MIL")
  print("Su salario total es: ",total)
  print("TRUE")

else:
  print("SALARIO INCORRECTO")
```
