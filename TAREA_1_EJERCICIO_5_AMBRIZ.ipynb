Este programa calcula el peso maximo de un pedido de dos productos, si excede el peso, te ddira por cuanto sobre pasa el limite

```
print("Para hacer un pedido, escriba lo siguiente")
print("NOTA; El pedido tiene un peso maximo de 10kg por caja")
a=float(input("Introduzca el numero de payasos que desea: "))
b=float(input("Introduzca el numero de juegos de mesa que desea: "))

x= a*0.150
y= b*0.170 

peso= x+y
exceso= peso-10

if peso >10:
   print("ERROR")
   print("El peso en kg sobre pasa por:",exceso)
else: 
   print("El peso en kg: ",peso) 
   print("EL PESO ES CORRECTO")

```
