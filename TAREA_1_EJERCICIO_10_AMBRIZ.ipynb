Este programa calcula el area de un circulo digitando el radio. Si lo desea, puede calcular el volumen de un cilindro escribiendo la altura

```
from math import pi

print("CALCULO DEL AREA SUPERFICIAL Y VOLUMEN DE UN CILINDRO")
r=int(input("Escribe el valor del radio: "))
h=int(input("Escribe el valor de la altura: "))

area= pi * r**2 

volumen= pi * r**2 * h

print("El area superficial del cilindro es: ", area)
print("El volumen del cilindro es: ", volumen)

```
