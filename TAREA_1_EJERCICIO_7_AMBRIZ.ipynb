Este programa calcula el precio de la entrada al museo, eligiendo una opcion para tu edad y tu ocupacion 

```
print("BIENVENIDOS AL MUSEO")
print(" Por favor seleccione una opcion dependiendo su edad")
x=int(input("(1- Menor de 4 años) (2- Entre 4 y 16 años) (3- Adulto) "))

print("Por favor seleccione una opcion segun su ocupacion")
y=int(input("(1-Alumno con credencial) (2-Docente con credencial) (3-INAPAM) "))

if x == 1:
  print("Menor de 4 años: GRATIS")

elif x == 2 and y == 1:
  costo = 50 * .10
  total = 50 - costo
  print("Entre 14 y 16 años, Estudiante con credencial, el total es: $", total)

elif x == 3 and y == 1:
  costo = 100 * .10
  total = 100 - costo
  print("Adulto, con crendencial de estudiante, el total es: $", total)

elif x == 1 and y == 2:
  print("Menor de 4 años: GRATIS")

elif x == 2 and y == 2:
  costo = 50 * .30
  total = 50 - costo
  print("Entre 4 y 16 años, Docente con credencial, el total es: $", total)

elif x == 3 and y == 2:
  costo = 100 * .30
  total = 100 - costo
  print("Adulto, Docente con credencial, el total es: $", total)

elif x == 1 and y == 3:
  print("Menor de 4 años: GRATIS")

elif x == 2 and y == 3:
  costo = 50 * .50
  total = 50 - costo
  print("Entre 4 y 16 años, con INAPAM, el total es: $", total)

elif x == 3 and y == 3:
  costo = 100 * .50
  total = 100 - costo
  print("Adulto, con INAPAM, el total es: $", total)

else:
  print("ERROR REINICIE PROGRAMA")
```
